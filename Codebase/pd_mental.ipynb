```python
import pandas as pd
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt 
```

```python
calls = pd.read_csv('/Users/gageholcomb/Desktop/Capstone/Data/Regional/Police Incidents/Metro_Nashville_Police_Department_Calls_for_Service_2023_5456239620798107499.csv')
```

```python
calls = calls.loc[calls['Tencode_Description'] == 'MENTALLY ILL PERSON']
```

```python
calls = calls[['Disposition_Description']]
```

```python
calls = calls.groupby(['Disposition_Description']).value_counts()
```

```python
calls = calls.reset_index()
```

```python
calls = calls.sort_values(by= ['count'], ascending = False)
```

```python
x = calls['Disposition_Description']
y = calls['count']
plt.bar(x, y)
plt.xlabel("Type of Mental Call Response - NPD 2023") 
plt.ylabel("No. of Responses")
plt.tick_params(axis='both', which='major', labelsize=10)
plt.tick_params(axis='both', which='minor', labelsize=8)
plt.xticks(fontsize=6, rotation=20)
```
