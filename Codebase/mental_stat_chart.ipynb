```python
import pandas as pd
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.arange(5) 
y1 = [31.4, 18.8, 16.7, 5.3, 2.7] 
y2 = [7.2, 4.6, 4.3, 1.9, 0.8] 
width = 0.4
  
# plot data in grouped manner of bar type 
plt.bar(x-0.2, y1, width) 
plt.bar(x+0.2, y2, width) 
plt.xticks(x, ['Any Mental Health Disorder', 'Depression', 'Anxiety', 'Depression & Anxiety', 'Other Episodic Mood Disorder']) 
plt.xlabel("Type of Mental Health Disorder") 
plt.ylabel("No. in Millions")
plt.legend(["Non-Opioid User", "Opioid User"]) 
plt.tick_params(axis='both', which='major', labelsize=10)
plt.tick_params(axis='both', which='minor', labelsize=8)
plt.xticks(fontsize=6, rotation=15)
```
