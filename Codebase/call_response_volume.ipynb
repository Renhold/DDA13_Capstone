```python
import pandas as pd
from calendar import month_name
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
multi = pd.read_csv('/Users/gageholcomb/Desktop/Capstone/Data/Regional/Mental Calls/Mental_Calls_by_Precinct_Plain_Text_Incident_Types_070121_022822.csv')
```

```python
multi = multi.rename(columns={'Assist Ems':'Assist EMS','Welfate Check':'Welfare Check'})
```

```python
multi = multi.drop(multi.loc[multi['Month'] == 'Total'].index)
```

```python
month_lookup = list(month_name)
multi.Month = sorted(multi.Month, key=month_lookup.index)
```

```python
multi = multi[['Month', 'Mental Condition', 'Assist EMS', 'Welfare Check']]
```

```python
plt.subplots(figsize=(8, 5))
sns.lineplot(x='Month', y='Mental Condition', data=multi, label='Mental Condition', color='r')
sns.lineplot(x='Month', y='Assist EMS', data=multi, label='Assist EMS')
sns.lineplot(x='Month', y='Welfare Check', data=multi, label='Welfare Check', color='g').set(title='Call Response Volume - NFD EMS 2020', xlabel='Month', ylabel='Call Volume by Type')
plt.margins(x=0)
sns.set_style('darkgrid')   
plt.margins(x=0)
plt.legend()
plt.show()
```
